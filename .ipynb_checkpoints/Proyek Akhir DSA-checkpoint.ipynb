```python
mode = film['tagline'].mode()
film['tagline'].fillna(str(mode), inplace = True)
```

```python
company = film['production_companies']
genre = film['genres']
prod_country = film['production_countries']

film = film.drop("homepage",1)
film = film.drop("keywords",1)
film = film.drop("original_language",1)
film = film.drop("original_title",1)
film = film.drop("overview",1)
film = film.drop("release_date",1)
film = film.drop("tagline",1)
film = film.drop("title",1)

ls = []
for i in company:
    v = json.loads(i)
    if len(v) != 0:
        ls.append(v[0].get('name'))
    else:
        ls.append('None')

film['production_companies'] = ls

ls = []
for i in genre:
    v = json.loads(i)
    if len(v) != 0:
        ls.append(v[0].get('name'))
    else:
        ls.append('None')

film['genres'] = ls

ls = []
for i in prod_country:
    v = json.loads(i)
    if len(v) != 0:
        ls.append(v[0].get('name'))
    else:
        ls.append('None')

film['production_countries'] = ls
```

```python
film.head()
```

```python
spoken = film['spoken_languages']

ls = []
for i in spoken:
    v = json.loads(i)
    if len(v) != 0:
        ls.append(v[0].get('name'))
    else:
        ls.append('None')

film['spoken_languages'] = ls
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
film_cols = film.columns
train_cols = [x for x in film_cols if x != 'revenue']

X = film[train_cols]
y = film['revenue']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=91, test_size=0.3,
)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)*100
print("Akurasi: ", accuracy,"%")
```

```python
confusion_matrix(y_test, y_pred)
```
